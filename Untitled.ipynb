```python
from datetime import datetime, timedelta
import pandas as pd
import matplotlib.pyplot as plt
from arch import arch_model
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import numpy as np
```

NIFTY50 Volatility


```python
df = pd.read_csv("NIFTY 50-12-11-2023-to-12-11-2024.csv")
```

```python
df.head()
```

```python
print(df.columns)
```

```python
df.columns = df.columns.str.strip()
df['Date'] = pd.to_datetime(df['Date'])
```

```python
returns = 100 * df['Close'].pct_change().dropna()
print(returns.head())
```

```python
plt.figure(figsize = (10, 4))
plt.plot(returns)
plt.ylabel("PCT returns", fontsize = 16)
plt.title('Nifty 50 retuns', fontsize = 20)
```

PACF

```python
plot_pacf(returns**2)
plt.show()
```

Fit GRACH(1, 1)

```python
model = arch_model(returns, p = 1, q = 1)
```

```python
model_fit = model.fit()
```

```python
rolling_predictions = []

testsize = 247

for i in range(testsize):
    train = returns[:-(testsize - i)]
    model = arch_model(train, p=1, q=1)
    model_fit = model.fit(disp="off")
    pred = model_fit.forecast(horizon=1)
    rolling_predictions.append(np.sqrt(pred.variance.values[-1, :][0]))

print("Rolling predictions:", rolling_predictions)
```

```python
rolling_predictions = pd.Series(rolling_predictions, index = returns.index[-247:])
```

```python
plt.figure(figsize = (10, 4))
true,  = plt.plot(returns[-247:])
preds, = plt.plot(rolling_predictions)
plt.title("Volatility Predictions", fontsize = 20)
plt.legend(["True, returns", "predicted volatility"], fontsize = 16)
```

```python
train = returns
model = arch_model(train, p = 1, q = 1)
model_fit = model.fit(disp = "off")
```

```python

df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df.set_index('Date', inplace=True)

last_date = df.index[-1]
last_date
```

```python
pred = model_fit.forecast(horizon = 20)
future_dates = [last_date + timedelta(days = i) for i in range(1, 21)]
pred = pd.Series(np.sqrt(pred.variance.values[-1,:]), index = future_dates)
```

```python
plt.figure(figsize = (15, 5))
plt.plot(pred)
plt.title("Volatility prediction - next 7 days", fontsize = 20)
```
